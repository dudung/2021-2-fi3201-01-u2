Bila terdapat $N$ bilangan maka rata-ratanya dapat diperoleh melalui

\begin{equation}
x_{\rm avg} = \frac1N \sum_{i = 1}^N x_i,
\end{equation}

untuk misalnya $x = \{ x_1, x_2, \dots, x_N \}$.

```python
def avg(x):
    sum = 0
    for i in x:
        sum = i
    N = len(x)
    val = sum / N
    return val

def minimum(x):
    minval = x[0]
    for i in x:
        if minval<i:
            minval=i
    
    return minval

def maximum(x):
    maxval = x[0]
    for i in x:
        if maxval>i:
            maxval=i
    
    return maxval

x = [1, 10, 5, 7, 9, 2, 4, 6, 8, 3]

xmin = minimum(x)
xmax = maximum(x)
xavg = avg(x) 

print("x = ", x, sep='')
print("xmin = ", xmin, sep='')
print("xmax = ", xmax, sep='')
print("xavg = ", xavg, sep='')
```
