```python
import math
import matplotlib.pyplot as plt
```

## function
Sebuah fungsi matematika, misalnya saja

<a name='eqn1'></a>
\begin{equation}\tag{1}
f(x) = \left[ x - 2.5 e^{-(x - 2.8)} \right] \sin \tfrac23 \pi (x-1)
\end{equation}

dapat digambarkan dengan dalam rentang $x \in [x_{\rm beg}, x_{\rm end}]$.

```python
# define a function of x
def f(x):
    y1 = (x -  2.5 * math.exp(-(x - 2.8)))
    y2 = math.sin(2 * math.pi * (x - 1) / 3)
    y = y1 * y2
    return y
```

```python
# create data
xbeg = 0
xend = 10
N = 100
dx = (xend - xbeg) / N

xx = []
yy = []

for i in range(0, N + 1):
    x = xbeg + i * dx
    y = f(x)
    
    xx.append(x)
    yy.append(y)
```

```python
# plot data
plt.ion()
fig, ax = plt.subplots()
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
plt.plot(xx, yy)

# draw horizontal line
line = plt.Line2D((xbeg, xend), (0, 0), color='#aaa', lw=1, ls='dashed')
plt.gca().add_line(line)

# show plot result
plt.show()
```

## another function
Terdapat fungsi lain berbentuk

<a name='eqn2'></a>
\begin{equation}\tag{2}
f(x) = e^{-ax} \sin bx
\end{equation}

dalam rentang $x \in [x_{\rm beg}, x_{\rm end}]$ yang memberikan hasil

![](draw_math_function.png)

dengan nilai-nilai $a$, $b$, $x_{\rm beg}$, $x_{\rm end}$, dan $N$ tertentu.

```python
# set parameters

a = 0.4
b = 5
xbeg = 0
xend = 5
N = 150

# define another function of x
def g(x):
    y = math.exp(-a * x) * math.sin(b * x)
    return y

# create data
dx = (xend - xbeg) / N

xx = []
yy = []

for i in range(0, N + 1):
    x = xbeg + i * dx
    y = g(x)
    
    xx.append(x)
    yy.append(y)

# plot data
plt.ion()
fig, ax = plt.subplots()
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
plt.plot(xx, yy)

# draw horizontal line
line = plt.Line2D((xbeg, xend), (0, 0), color='#aaa', lw=1, ls='dashed')
plt.gca().add_line(line)
line = plt.Line2D((xbeg, xend), (0.4, 0.4), color='#aaa', lw=1, ls='dashed')
plt.gca().add_line(line)
line = plt.Line2D((xbeg, xend), (0.8, 0.8), color='#aaa', lw=1, ls='dashed')
plt.gca().add_line(line)
line = plt.Line2D((xbeg, xend), (-0.4, -0.4), color='#aaa', lw=1, ls='dashed')
plt.gca().add_line(line)

# show plot result
plt.show()

# print parameters
print("a = ", a, sep='')
print("b = ", b, sep='')
print("xbeg = ", xbeg, sep='')
print("xend = ", xend, sep='')
print("N = ", N, sep='')
```
